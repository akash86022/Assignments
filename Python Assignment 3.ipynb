**Custom Myreduce() function**

```
def myreduce(func,lst):
  a=lst[0]
  for x in lst[1:]:
    a=func(a,x)
  return a
```

```
lst=[2,2,3,4]
```

```
res=myreduce(lambda a,b:a*b,lst)
```

```
print(res)
```

**Custom myfilter() function**

```
def myfilter(func,iterable):
  a=[]
  if func is None:   ### Handling the case if None is Passed, Function filters out False/0 values
    for i in iterable:
      if i!=False:
        a.append(i)
  else:      
    for i in iterable:
      if func(i)==True:
        a.append(i)
  return a
```

```
# list of alphabets
alphabets = ['a', 'b', 'd',False, 'e', 'i', 'j', 'o',0]

#Sample function that filters vowels
def filterVowels(alphabet):
    vowels = ['a', 'e', 'i', 'o', 'u']

    if(alphabet in vowels):
        return True
    else:
        return False
```

```
r=myfilter(filterVowels,alphabets)
s=myfilter(None,alphabets)
```

```
print('r:',r ," ",'s:',s)
```

## **List Comprehension**

```
print([ a*b for a in ['x','y','z'] for b in range(1,5)]) ##  List Comprehension 1
```

```
print([(i*j) for j in range(1,5) for i in ['x','y','z']])  ## List Comprehension 2
```

```
print([[i+j] for j in range(1,4) for i in range(1,4)])  ## List Comprehension 3
```

```
print([[i+j for j in range(1,5)] for i in range(1,5)])  ## List Comprehension 4
```

```
print([(i,j) for j in range(1,4) for i in range(1,4)])  ## List Comprehension 5
```
