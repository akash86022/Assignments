```
import pandas as pd
import numpy as np
```

```
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```
df
```

```
df.FlightNumber.interpolate(inplace=True) # fixing Flight number values
df.FlightNumber.astype('int64')
```

```
new=pd.DataFrame(df.From_To.str.split('_',expand=True)) #splitting From_To column
```

```
df=pd.concat([df,new],axis=1)
```

```
df.rename(columns={0:'From',1:'To'},inplace=True)
```

```
df.drop(labels='From_To',axis=1,inplace=True) #Dropping From_To column
```
