```
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
from sklearn import decomposition 
from sklearn import datasets 
import pandas as pd
%matplotlib inline
```

```
i=datasets.load_iris()
```

Data Preprocessing

```
iris=pd.DataFrame(i['data'],columns=i['feature_names'])
tgt=pd.DataFrame(i['target'],columns=['output'])
iris_final=pd.concat([iris,tgt],axis=1)
```

```
tgt_names={}
y=[x for x in i.target_names]
for j in range (0,3):
  tgt_names[j]=y[j]
```

```
iris_final['output']=iris_final['output'].map(tgt_names)
```
